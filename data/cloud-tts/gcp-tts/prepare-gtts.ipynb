```python
import json
    
with open('../malaya/postprocessing-edge-tts-news.json') as fopen:
    text = json.load(fopen)
```

```python
lens = 0
filtered = []
for t in text:
    l = len(t[1])
    filtered.append(t[1])
    
    lens += l
    
    if lens >= 3000000:
        break
```

```python
len(filtered), len(text)
```

```python
with open('../malaya/synthetic-data-tts.json') as fopen:
    text = json.load(fopen)
    
for s in text:
    if s.count(' ') / len(s) > 0.2:
        filtered.append(s)
```

```python
filtered[-100:]
```

```python
with open('gtts-text.json', 'w') as fopen:
    json.dump(filtered, fopen)
```
