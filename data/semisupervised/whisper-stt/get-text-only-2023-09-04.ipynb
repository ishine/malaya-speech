```python
import string
printable = set(string.printable)
```

```python
ms = open('ms-2023-09-04.jsonl', 'w')
en = open('en-2023-09-04.jsonl', 'w')
```

```python
''.join([s.text for s in data[0]['asr_model'][2]])
```

```python
from tqdm import tqdm
import json

for f in tqdm(files):
    with open(f, 'rb') as fopen:
        data = pickle.load(fopen)
    ms_text, en_text = [], []
    for d in data:
        
        ms_ = ''.join([s.text for s in d['asr_model'][1]])
        en_ = ''.join([s.text for s in d['asr_model'][2]])
        
        if not len(set(ms_) - printable):
            ms_text.append(ms_)
        
        if not len(set(en_) - printable):
            en_text.append(en_)
        
    data = {
        'file': f,
        'ms_text': ''.join(ms_text),
    }
    
    ms.write(f'{json.dumps(data)}\n')
    
    data = {
        'file': f,
        'en_text': ''.join(en_text),
    }
    
    en.write(f'{json.dumps(data)}\n')
```

```python
ms.close()
en.close()
```
