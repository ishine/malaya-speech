```python
from dia.model import Dia

model = Dia.from_pretrained("mesolitica/Malaysian-Dia-1.6B", compute_dtype="float16")
```

```python
clone_from_text = "[S1] Uhm, hello, selamat pagi ye, saya dari customer service, boleh saya bantu you dengan apa apa ke?"
clone_from_audio = "husein-assistant-trim.mp3"
```

```python
text_to_generate = "[S1] Terdapat beberapa sebab mengapa kerajaan Malaysia teragak-agak untuk menangani isu ini dengan berkesan. o"
output = model.generate(
    clone_from_text + text_to_generate, audio_prompt=clone_from_audio, use_torch_compile=True, verbose=True, temperature=1.2,
)
```

```python
import IPython.display as ipd
ipd.Audio(output[:-1024], rate = 44100)
```
