```python
from dia.model import Dia

model = Dia.from_pretrained("mesolitica/Malaysian-Podcast-Dia-1.6B", compute_dtype="float16")
```

```python
clone_from_text = "[S1] Uhm, hello, selamat pagi ye, saya dari customer service, boleh saya bantu you dengan apa apa ke?"
clone_from_audio = "husein-assistant.mp3"
```

```python
text_to_generate = "[S1] Uhm, yes, sekejap ya, puan, uhm, saya akan buat semakan, uhm, melalui sistem yang saya ada. Hahahaha, sementara itu, do you have life insurance? if not, i have insurance Kasih suitable for you. o"
output = model.generate(
    clone_from_text + text_to_generate, audio_prompt=clone_from_audio, use_torch_compile=True, verbose=True, temperature=1.2,
)
```

```python
import IPython.display as ipd
ipd.Audio(output, rate = 44100)
```

```python
text_to_generate = "[S1] Terima kasih puan kerana menghubungi perkhidmatan pelanggan Hotel Husein. Uhm, puan ingin tahu tentang kemudahan WiFi ataupun masalah air panas, uhm, boleh saya tahu tempoh menginap puan untuk semakan selanjutnya? o"
output = model.generate(
    clone_from_text + text_to_generate, audio_prompt=clone_from_audio, use_torch_compile=True, verbose=True, temperature=1.2,
)
ipd.Audio(output, rate = 44100)
```
