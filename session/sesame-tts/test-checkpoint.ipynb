```python
from generator import Segment, load_csm_1b
import torch
import torchaudio

device = "cuda" if torch.cuda.is_available() else "cpu"
```

```python
generator = load_csm_1b(model_name = 'mesolitica/Malaysian-sesame-csm-1b', device=device)
```

```python
def prepare_prompt(text: str, speaker: int, audio_path: str) -> Segment:
    audio_tensor, _ = load_prompt_audio(audio_path)
    return Segment(text=text, speaker=speaker, audio=audio_tensor)

def load_prompt_audio(audio_path: str) -> torch.Tensor:
    audio_tensor, sample_rate = torchaudio.load(audio_path)
    if audio_tensor.shape[0] != 1:
        audio_tensor = audio_tensor.mean(dim=0)
    audio_tensor = audio_tensor.squeeze(0)
    if sample_rate != generator.sample_rate:
        audio_tensor = torchaudio.functional.resample(
            audio_tensor, orig_freq=sample_rate, new_freq=generator.sample_rate
        )
    return audio_tensor, generator.sample_rate
```

```python
text_prompt_speaker = 'Saya ingin nyatakan bahawa, cara menangani masalah ekonomi dalam krisis, tidak sama dalam keadaan biasa, sebab itu soal defisit, soal hutang, soal termasuk moratorium.'
audio_prompt_speaker = 'anwar-ibrahim.wav'
audio_prompt = prepare_prompt(text_prompt_speaker, 0, audio_prompt_speaker)
prompt_segments = [audio_prompt]
```

```python
line = 'Husein bin Zolkepli sebenarnya, sangat busuk, tak mandi, ketiak pun masham, but he is comel.'
```

```python
audio_tensor, tokens = generator.generate(
    text=line,
    speaker=0,
    context=prompt_segments,
    max_audio_length_ms=30_000, temperature = 0.3
)
```

```python
import IPython.display as ipd
ipd.Audio(audio_tensor.cpu().numpy(), rate = generator.sample_rate)
```
