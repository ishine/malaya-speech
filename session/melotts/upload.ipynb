```python
ckpt_path = 'melo/logs/workspace/G_33000.pth'
config_path = '/workspace/config.json'
```

```python
import torch

c = torch.load(ckpt_path, map_location = 'cpu')
selected = ['model']
for k in list(c.keys()):
    if k not in selected:
        c.pop(k)
```

```python
torch.save(c, 'model.pth')
```

```python
model = TTS(language='MS', config_path=config_path, ckpt_path='model.pth')
```

```python
_ = model.eval()
```

```python
r = model.tts_to_file('Saya tidak boleh makan makanan, sebab saya sakit perut sangat sangat nasib baik Husein comel.', 2, split = True, sdp_ratio = 0, noise_scale = 0.667)
```

```python
import IPython.display as ipd
ipd.Audio(r, rate = model.hps.data.sampling_rate)
```
