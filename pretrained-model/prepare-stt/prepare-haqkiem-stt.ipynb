```python
import pandas as pd
from glob import glob
from tqdm import tqdm
```

```python
df = pd.read_csv(f'haqkiem/metadata.csv', header = None, sep = '|')
txts = df.values.tolist()
haqkiem = []
for f in tqdm(txts):
    text = f[1]
    text = text.split('.,,')[0]
    f = f[0]
    r = f'haqkiem/{f}.wav'
    
    haqkiem.append({
        'new_text': text,
        'audio_filename': r
    })
```
